```python
from hashlib import md5
from time import time
from string import printable
from itertools import product, count


def hashes(encoding):
    chars = [c.encode(encoding) for c in printable]
    for length in count(start=1):
        for pwd in product(chars, repeat=length):
            yield b''.join(pwd)


def crack_hash(search_hash, encoding):
    for pwd in hashes(encoding):
        if md5(pwd).digest() == search_hash:
            return pwd.decode(encoding)


if __name__ == "__main__":
    encoding = 'ascii'  # utf-8 
    hashInf = hex(bit8randomNumber) #hex from task 2
    password_hash = md5(hashInf.encode(encoding)).digest()

    start = time()
    cracked_hash = crack_hash(password_hash, encoding)
    end = time()
    print(f"Hash found: {cracked_hash}")
    print(f"Time: {end - start} seconds.")
```
